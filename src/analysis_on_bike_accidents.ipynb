```python
import matplotlib.pyplot as plt

import fetch_data as fd
```

```python
REPORT_YEARS = [2021, 2022, 2023, 2024]
report_dict = fd.get_dfs(REPORT_YEARS)
```

Total bike accidents


```python
bike_accident_totals: dict[int, int] = {}

for year in REPORT_YEARS:
    df_yr = report_dict[year]
    total_acc = df_yr.shape[0]
    bike_accident_totals[year] = df_yr[(df_yr["IstRad"] == 1)].shape[0]
```

```python
plt.plot(bike_accident_totals.keys(), bike_accident_totals.values(), marker="o")  # type: ignore
plt.title("Proportion of Bike Accidents Over Time")  # type: ignore
plt.xlabel("Year")  # type: ignore
plt.ylabel("Proportion of Bike Accidents")  # type: ignore
plt.xlim(2020.5, 2024.5)  # type: ignore
plt.xticks(list(bike_accident_totals.keys()))  # type: ignore
plt.show()  # type: ignore
```

Plot the proportion of bike accidents.


```python
bike_accident_prop: dict[int, float] = {}

for year in REPORT_YEARS:
    df_yr = report_dict[year]
    total_acc = df_yr.shape[0]
    bike_acc = df_yr[(df_yr["IstRad"] == 1)].shape[0]
    bike_accident_prop[year] = bike_acc / total_acc if total_acc > 0 else 0
```

```python
plt.plot(bike_accident_prop.keys(), bike_accident_prop.values(), marker="o")  # type: ignore
plt.title("Proportion of Bike Accidents Over Time")  # type: ignore
plt.xlabel("Year")  # type: ignore
plt.ylabel("Proportion of Bike Accidents")  # type: ignore
plt.ylim(0.3, 0.35)  # type: ignore
plt.xlim(2020.5, 2024.5)  # type: ignore
plt.xticks(list(bike_accident_prop.keys()))  # type: ignore
plt.show()  # type: ignore
```
